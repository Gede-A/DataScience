```python
import pandas as pd

# Load the dataset
data = pd.read_csv('Titanic-Dataset.csv')
data.head()
```

```python
# Check for missing values
data.isnull().sum()

# Fill missing values in 'Age' with the median
data['Age'].fillna(data['Age'].median(), inplace=True)

# Fill missing values in 'Embarked' with the mode
data['Embarked'].fillna(data['Embarked'].mode()[0], inplace=True)

# Drop the 'Cabin' column due to too many missing values
data.drop(columns=['Cabin'], inplace=True)

# Convert 'Sex' and 'Embarked' to categorical type
data['Sex'] = data['Sex'].astype('category')
data['Embarked'] = data['Embarked'].astype('category')

data.info()
```

```python
# Summary statistics
data.describe()

# Count of survivors and non-survivors
data['Survived'].value_counts()

# Group by 'Sex' and 'Survived' to see survival rate by gender
data.groupby(['Sex', 'Survived']).size().unstack()

# Group by 'Pclass' to see survival rate by passenger class
data.groupby(['Pclass', 'Survived']).size().unstack()
```

```python
# Ensure 'Survived' and 'Pclass' are treated as categorical
data['Survived'] = data['Survived'].astype('category')
data['Pclass'] = data['Pclass'].astype('category')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Sample data for demonstration
data = pd.DataFrame({
    'Age': [22, 38, 26, 35, 35],
    'Survived': [1, 1, 0, 1, 0],
    'Sex': ['male', 'female', 'female', 'female', 'male'],
    'Pclass': [3, 1, 3, 1, 3],
    'Fare': [7.25, 71.2833, 7.925, 53.1, 8.05]
})

# Convert categorical columns to string type
data['Survived'] = data['Survived'].astype(str)
data['Pclass'] = data['Pclass'].astype(str)

# Set the aesthetic style of the plots
sns.set_style('whitegrid')

# Histogram of ages
plt.figure(figsize=(10, 6))
sns.histplot(data['Age'], bins=30, kde=True)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

# Survival count by gender
plt.figure(figsize=(10, 6))
sns.countplot(x='Survived', hue='Sex', data=data)
plt.title('Survival Count by Gender')
plt.xlabel('Survived')
plt.ylabel('Count')
plt.show()

# Survival rate by passenger class
plt.figure(figsize=(10, 6))
sns.countplot(x='Pclass', hue='Survived', data=data)
plt.title('Survival Rate by Passenger Class')
plt.xlabel('Pclass')
plt.ylabel('Count')
plt.show()

# Pair plot to see pairwise relationships
sns.pairplot(data, hue='Survived', vars=['Age', 'Fare'])
plt.show()
```

```python
data['Survived'] = pd.to_numeric(data['Survived'], errors='coerce')
data['Pclass'] = pd.to_numeric(data['Pclass'], errors='coerce')
data['Fare'] = pd.to_numeric(data['Fare'], errors='coerce')
data['Age'] = pd.to_numeric(data['Age'], errors='coerce')
```

```python
print(data.info())
```

```python
# One-hot encode the 'AgeGroup' column
age_group_encoded = pd.get_dummies(data['AgeGroup'], prefix='AgeGroup')

# Concatenate the one-hot encoded columns with the original dataframe
data_encoded = pd.concat([data, age_group_encoded], axis=1)

# Drop the original 'AgeGroup' column
data_encoded.drop(columns=['AgeGroup'], inplace=True)

# Calculate correlation matrix
corr_matrix = data_encoded.corr()

# Plot the correlation matrix
plt.figure(figsize=(10, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
